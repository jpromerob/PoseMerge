```python
from matplotlib import pyplot as mp
import numpy as np

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

x_values = np.linspace(-10, 10, 500)
for mu, sig in [(-1, 1), (0, 2), (2, 3)]:
    mp.plot(x_values, gaussian(x_values, mu, sig))

mp.show()
```

```python

from matplotlib import pyplot as plt
import numpy as np

from scipy import stats

'''
Parameter 'k' should be smaller than 1 (usually 0.1 or less)
'''
def get_errors_from_snn(nb_pts, k):

    max_val = 1
    min_val = 0
    abs_max_err = (max_val-min_val)*k
    
    array_true = np.random.uniform(low=0.0, high=1.0, size=(nb_pts,))
    array_snn = np.zeros(nb_pts)
    
    for idx in range(len(array_true)):
        err = np.random.uniform(low=-abs_max_err, high=abs_max_err, size=(1,))
        array_snn[idx] = array_true[idx]*(1 + err)


    error = array_true-array_snn

    std = np.std(error)
    mean = np.mean(error)
    
    return mean, std


nb_pts = 1000

cam_pdf_params = np.zeros((3,3,2)) # {1,2,3} | {x,y,z} | {mean, std}

#Cam 1
cam_pdf_params[0,0,:] = get_errors_from_snn(nb_pts, 0.05)
cam_pdf_params[0,1,:] = get_errors_from_snn(nb_pts, 0.05)
cam_pdf_params[0,2,:] = get_errors_from_snn(nb_pts, 0.20)

#Cam 2
cam_pdf_params[1,0,:] = get_errors_from_snn(nb_pts, 0.04)
cam_pdf_params[1,1,:] = get_errors_from_snn(nb_pts, 0.06)
cam_pdf_params[1,2,:] = get_errors_from_snn(nb_pts, 0.19)

#Cam 3
cam_pdf_params[2,0,:] = get_errors_from_snn(nb_pts, 0.06)
cam_pdf_params[2,1,:] = get_errors_from_snn(nb_pts, 0.04)
cam_pdf_params[2,2,:] = get_errors_from_snn(nb_pts, 0.21)


xyz = np.linspace(norm.ppf(0.40), norm.ppf(0.60), 1000)
for i in range(3): # 1,2,3
    for j in range(3): # x,y,z 
        
        ## y-axis as the gaussian
        pdf = stats.norm.pdf(xyz, cam_pdf_params[i,j,0], cam_pdf_params[i,j,1])

        ## plot data
        plt.plot(xyz, pdf)


```

```python

def get_mixture(pdf_1, pdf_2, pdf_3):
    
    scaler = 0.33
    pdf_mixt = (pdf_1 + pdf_2 + pdf_3) * scaler
    
    return pdf_mixt


def get_combination(pdf_1, pdf_2, pdf_3):
    pdf_comb = (pdf_1 * pdf_2 * pdf_3) 
    pdf_comb = ((pdf_1.sum() + pdf_2.sum() + pdf_3.sum()) / 3.0) * (pdf_comb.astype(float) / pdf_comb.astype(float).sum())
    
    return pdf_comb



xyz = np.linspace(0.4,0.6, 1000)
snn_x_values = [0.465, 0.511, 0.565]

pdf_1 = stats.norm.pdf(xyz, snn_x_values[0]+cam_pdf_params[0,0,0], cam_pdf_params[0,0,1])  # Cam 1
pdf_2 = stats.norm.pdf(xyz, snn_x_values[1]+cam_pdf_params[1,0,0], cam_pdf_params[1,0,1])  # Cam 2
pdf_3 = stats.norm.pdf(xyz, snn_x_values[2]+cam_pdf_params[2,0,0], cam_pdf_params[2,0,1])  # Cam 3
pdf_comb = get_combination(pdf_1, pdf_2, pdf_3)
pdf_mixt = get_mixture(pdf_1, pdf_2, pdf_3)

plt.plot(xyz, pdf_1)
plt.plot(xyz, pdf_2)
plt.plot(xyz, pdf_3)
plt.plot(xyz, pdf_comb)
plt.plot(xyz, pdf_mixt)
```

```python

# Importing the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
 
 
plt.style.use('seaborn-dark')
plt.rcParams['figure.figsize']=14,6
fig = plt.figure()
 
# Initializing the random seed
random_seed=1000
 
# List containing the variance
# covariance values
cov_val = [-0.8, 0, 0.8]
 
# Setting mean of the distributino
# to be at (0,0)
mean = np.array([3,0])
 
# Storing density function values for
# further analysis
pdf_list = []
 
# Iterating over different covariance values
for idx, val in enumerate(cov_val):
     
    # Initializing the covariance matrix
    cov = np.array([[1, val], [val, 1]])
     
    # Generating a Gaussian bivariate distribution
    # with given mean and covariance matrix
    distr = multivariate_normal(cov = cov, mean = mean,
                                seed = random_seed)
     
    # Generating a meshgrid complacent with
    # the 3-sigma boundary
    mean_1, mean_2 = mean[0], mean[1]
    sigma_1, sigma_2 = cov[0,0], cov[1,1]
     
    x = np.linspace(-3*sigma_1+mean_1, 3*sigma_1+mean_1, num=100)
    y = np.linspace(-3*sigma_2+mean_2, 3*sigma_2+mean_2, num=100)
    X, Y = np.meshgrid(x,y)
     
    # Generating the density function
    # for each point in the meshgrid
    pdf = np.zeros(X.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            pdf[i,j] = distr.pdf([X[i,j], Y[i,j]])
     
    # Plotting the density function values
    key = 131+idx
    ax = fig.add_subplot(key, projection = '3d')
    ax.plot_surface(X, Y, pdf, cmap = 'viridis')
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(f'Covariance between x1 and x2 = {val}')
    pdf_list.append(pdf)
    ax.axes.zaxis.set_ticks([])
 
plt.tight_layout()
plt.show()
 
# Plotting contour plots
for idx, val in enumerate(pdf_list):
    plt.subplot(1,3,idx+1)
    plt.contourf(X, Y, val, cmap='viridis')
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(f'Covariance between x1 and x2 = {cov_val[idx]}')
plt.tight_layout()
plt.show()
```

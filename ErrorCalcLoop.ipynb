```python

%matplotlib notebook

import numpy as np
from multitools import get_w_bases
from multitools import create_mgd, prepare_ground
from multitools import predict_pose
from multitools import visualize_intersection, visualize_all_cigars



```

```python
# Number of points, per axis, to evaluate
nb_pts = 101

nb_s_y = 9
nb_s_xz = int(nb_s_y/2)+1
w_x = np.linspace(-0.5, 0, num=nb_s_xz) 
w_y = np.linspace(0.0, 1, num=nb_s_y)
w_z = np.linspace(0.7, 1.2, num=nb_s_xz)

e_d = np.zeros((nb_s_xz, nb_s_y, nb_s_xz)) # error in euler distance in [cm]
count = 0
for i in range(nb_s_xz):
    for j in range(nb_s_y):
        for k in range(nb_s_xz):

            count += 1
            print("Sim {:.0f}/{:.0f}".format(count, nb_s_xz*nb_s_y*nb_s_xz))

            # Object's pose in world space
            w_pose = np.array([w_x[i], w_y[j], w_z[k]])  

            print("For w_pose = ({:.3f}, {:.3f}, {:.3f})".format(w_pose[0], w_pose[1], w_pose[2]))


            # Mean array and covariance matrix in virtual camera space
            μ = np.array([0,0,-0.6])
            Σ = np.array([[0.01,0,0],[0,0.01,0],[0,0,1.2]])    

            # Go to virtual space and get data (matrices) in the process
            r_trl, r2w, v2r, r_obj_poses, v_obj_poses, r_cam_poses = prepare_ground(w_pose)

            # Create gaussians in virtual space and transform them into world space
            new_μ, rv = create_mgd(μ, Σ, r_trl, r2w, v2r, v_obj_poses)

            # Do predictions
            x, y, z, p, idx, prediction, e_d[i,j,k] = predict_pose(nb_pts, rv, new_μ,  w_pose)
            
filename = "e_d" + str(nb_s_xz) + "_" + str(nb_s_y) + "_" + str(nb_s_xz) + ".npy"
print(filename)
np.save(filename,e_d)
```

```python

e_d = 100*np.load(filename)
print(e_d.shape)

in_l_y = 1
in_l_xz = 1
in_y = [0+in_l_y,(nb_s_y-1)-in_l_y]
in_xz = [0+in_l_xz,(nb_s_xz-1)-in_l_xz]

if in_y[0] >= in_y[1]:
    error_sample = e_d
else:
    
    print("Workspace:")
    if in_xz[0] == in_xz[1]:
        error_sample = e_d[in_xz[0],in_y[0]:in_y[1],in_xz[0]]
        print("\t x in [{:.3f}]".format(w_x[in_xz[0]]))
        print("\t y in [{:.3f}, {:.3f}]".format(w_y[in_y[0]], w_y[in_y[1]]))
        print("\t z in [{:.3f}]".format(w_z[in_xz[0]]))
    else:    
        error_sample = e_d[in_xz[0]:in_xz[1],in_y[0]:in_y[1],in_xz[0]:in_xz[1]]

        print("\t x in [{:.3f}, {:.3f}]".format(w_x[in_xz[0]], w_x[in_xz[1]]))
        print("\t y in [{:.3f}, {:.3f}]".format(w_y[in_y[0]], w_y[in_y[1]]))
        print("\t z in [{:.3f}, {:.3f}]".format(w_z[in_xz[0]], w_z[in_xz[1]]))

print("Max error : {:.3f} [cm]".format(np.amax(error_sample)))
print("Mean error : {:.3f} [cm]".format(np.mean(error_sample)))
```

```python
import matplotlib.pyplot as plt
# Creating figure
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
nb_pts = 100
# 'Plot' workspace
xs = np.linspace(-0.7, 0, nb_pts)
ys = np.linspace(-0.1, 1, nb_pts)
zs = np.linspace(0, 1.6, nb_pts)

X, Y = np.meshgrid(xs, ys, indexing='ij')
Z = np.ones((nb_pts,nb_pts))*1.5
ax.plot_surface(X, Y, Z, alpha=0.2, color='k')

Y, Z = np.meshgrid(ys, zs, indexing='ij')
X = np.ones((nb_pts,nb_pts))*-0.6
ax.plot_surface(X, Y, Z, alpha=0.2, color='k')

X, Z = np.meshgrid(xs, zs, indexing='ij')
Y = np.ones((nb_pts,nb_pts))*0
ax.plot_surface(X, Y, Z, alpha=0.2, color='k')
    
xx, yy, zz = np.meshgrid(w_x, w_y, w_z, indexing='ij')

e_d = e_d

e_array = [16, 8, 4, 2, 1]
c_array = ['k', 'b', 'g', 'r', 'y']
for k in range(len(e_array)):   
    
    ai = (len(e_array)-k)/max(e_array)
    idx = e_d < e_array[k]
    if np.count_nonzero(idx) > 0:
        ax.scatter3D(xx[idx],yy[idx], zz[idx], e_d[idx], vmin=0, vmax=10, marker=".", alpha=ai, color=c_array[k])
    else:
        print("No pts closer than {:.3f} [cm]".format(e_array[k]))



print(xx.shape)
print(yy.shape)
print(zz.shape)
print(e_d.shape)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')    

ax.set_xlim([-1, 0.5])
ax.set_ylim([-0.5, 1.0])
ax.set_zlim([0, 1.5])
ax.view_init(-30, 10)    

plt.show()  
```
